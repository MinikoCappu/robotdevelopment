```python
import pandas as pd

df = pd.read_csv("третья.csv", encoding='utf-8')
df['Стоимость'] = df['Количество'] * df['Цена']

import xlsxwriter
workbook = xlsxwriter.Workbook('chart_bar.xlsx')
worksheet = workbook.add_worksheet()
worksheet.write_row('A1', ["Стоимость", "Название"])
worksheet.write_column('A2', df['Стоимость'])
worksheet.write_column('B2', df['Название'])
chart1 = workbook.add_chart({'type': 'bar'})


workbook.close()
```
