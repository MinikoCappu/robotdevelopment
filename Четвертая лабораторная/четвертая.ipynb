```python
import sqlite3
import random
from datetime import datetime, timedelta

conn = sqlite3.connect("data.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS Employees (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    full_name TEXT NOT NULL,
    department TEXT NOT NULL,
    salary REAL NOT NULL,
    hire_date TEXT NOT NULL
)
""")
first_names = [
    "Иван", "Пётр", "Алексей", "Сергей", "Андрей",
    "Дмитрий", "Михаил", "Николай", "Владимир", "Евгений"
]
last_names = [
    "Иванов", "Петров", "Сидоров", "Кузнецов", "Смирнов",
    "Попов", "Соколов", "Лебедев", "Козлов", "Новиков"
]
departments = ["IT", "HR", "Finance", "Logistics", "Marketing"]

for _ in range(120):
    full_name = f"{random.choice(last_names)} {random.choice(first_names)}"
    department = random.choice(departments)
    salary = random.randint(40000, 120000)

    hire_date = datetime.now() - timedelta(days=random.randint(100, 3000))
    hire_date = hire_date.strftime("%Y-%m-%d")

    cursor.execute("""
    INSERT INTO Employees (full_name, department, salary, hire_date)
    VALUES (?, ?, ?, ?)
    """, (full_name, department, salary, hire_date))

conn.commit()
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("data.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS logistics_bonus (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    employee_id INTEGER,
    full_name TEXT,
    salary REAL,
    bonus REAL
)
""")

cursor.execute("DELETE FROM logistics_bonus")

cursor.execute("""
SELECT id, full_name, salary
FROM Employees
WHERE department = 'Logistics'
""")

employees = cursor.fetchall()

for emp_id, full_name, salary in employees:
    bonus = salary * 0.15

    cursor.execute("""
    INSERT INTO logistics_bonus (employee_id, full_name, salary, bonus)
    VALUES (?, ?, ?, ?)
    """, (emp_id, full_name, salary, bonus))

conn.commit()
conn.close()
```
